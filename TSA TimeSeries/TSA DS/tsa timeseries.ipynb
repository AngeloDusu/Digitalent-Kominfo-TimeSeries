TRAINING & TESTING

MODEL ARIMA

```python
# Split the data into training and testing sets
train_size = int(len(data_2) * 0.8)
train, test = data_2[:train_size], data_2[train_size:]

# Extract the target variable for modeling
train_ketinggian = train['features__properties__ketinggian']
test_ketinggian = test['features__properties__ketinggian']
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np

# Build and fit the ARIMA model
arima_model = ARIMA(train_ketinggian, order=(5,1,0))
arima_fit = arima_model.fit()

# Make predictions on the test set
arima_predictions = arima_fit.forecast(steps=len(test_ketinggian))

# Calculate the RMSE for the ARIMA model
arima_rmse = np.sqrt(mean_squared_error(test_ketinggian, arima_predictions))
print(f'ARIMA RMSE: {arima_rmse}')
```

MODEL LSTM

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
train_ketinggian_scaled = scaler.fit_transform(train_ketinggian.values.reshape(-1, 1))
test_ketinggian_scaled = scaler.transform(test_ketinggian.values.reshape(-1, 1))

# Prepare the data for LSTM
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

time_step = 10
X_train, y_train = create_dataset(train_ketinggian_scaled, time_step)
X_test, y_test = create_dataset(test_ketinggian_scaled, time_step)

# Reshape input to be [samples, time steps, features] which is required for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32, verbose=1)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transform predictions to original scale
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate RMSE for the LSTM model
lstm_rmse = np.sqrt(mean_squared_error(y_test, test_predict))
print(f'LSTM RMSE: {lstm_rmse}')
```

Fungsi ini:

Menyaring data berdasarkan ID pintu air (jika diberikan).

Menetapkan datetime sebagai indeks dan mengurutkan data berdasarkan indeks ini.
Menyimpan hanya kolom numerik.
Melakukan resampling data untuk membuat data memiliki interval per jam.
Mengisi nilai yang hilang dengan metode forward fill.
Menyelaraskan data menggunakan MinMaxScaler.

"""


kesimpulan code di bawah

Kodingan tersebut menggunakan model LSTM untuk memprediksi ketinggian air berdasarkan data historis. Data diproses, dibuat urutan, dibagi menjadi set pelatihan dan pengujian, dan model dilatih serta diuji. Prediksi dibuat dan dibandingkan dengan nilai aktual untuk evaluasi model.

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load the data
df = banjir.copy()

# Function to preprocess data
def preprocess_data(df, water_gate_ids=None):
    if water_gate_ids:
        df = df[df['features__properties__pintu_air__id'].isin(water_gate_ids)]
    
    df['datetime'] = pd.to_datetime(df['features__properties__tanggal'] + ' ' + df['features__properties__jam'])
    df.set_index('datetime', inplace=True)
    df.sort_index(inplace=True)
    
    # Keep only numeric columns
    df_numeric = df.select_dtypes(include=[np.number])
    
    df_resampled = df_numeric.resample('H').mean()
    df_resampled.fillna(method='ffill', inplace=True)
    
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df_resampled[['features__properties__ketinggian']])
    scaled_df = pd.DataFrame(scaled_data, index=df_resampled.index, columns=['features__properties__ketinggian'])
    
    return scaled_df, scaler



# Preprocess data for specific water gates
manggarai_bkb_df, manggarai_bkb_scaler = preprocess_data(df, water_gate_ids=[6])
istiqlal_df, istiqlal_scaler = preprocess_data(df, water_gate_ids=[19])

# Preprocess data for overall prediction
overall_df, overall_scaler = preprocess_data(df)

# Function to create sequences
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i+seq_length]
        y = data[i+seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 24
X_manggarai, y_manggarai = create_sequences(manggarai_bkb_df.values, seq_length)
X_istiqlal, y_istiqlal = create_sequences(istiqlal_df.values, seq_length)
X_overall, y_overall = create_sequences(overall_df.values, seq_length)

# Split the data into training and testing sets
split_manggarai = int(len(X_manggarai) * 0.8)
X_manggarai_train, X_manggarai_test = X_manggarai[:split_manggarai], X_manggarai[split_manggarai:]
y_manggarai_train, y_manggarai_test = y_manggarai[:split_manggarai], y_manggarai[split_manggarai:]

split_istiqlal = int(len(X_istiqlal) * 0.8)
X_istiqlal_train, X_istiqlal_test = X_istiqlal[:split_istiqlal], X_istiqlal[split_istiqlal:]
y_istiqlal_train, y_istiqlal_test = y_istiqlal[:split_istiqlal], y_istiqlal[split_istiqlal:]

split_overall = int(len(X_overall) * 0.8)
X_overall_train, X_overall_test = X_overall[:split_overall], X_overall[split_overall:]
y_overall_train, y_overall_test = y_overall[:split_overall], y_overall[split_overall:]

# Build the LSTM model
def build_model():
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(seq_length, 1)))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train models for each dataset
model_manggarai = build_model()
model_manggarai.fit(X_manggarai_train, y_manggarai_train, batch_size=32, epochs=10, validation_data=(X_manggarai_test, y_manggarai_test))

model_istiqlal = build_model()
model_istiqlal.fit(X_istiqlal_train, y_istiqlal_train, batch_size=32, epochs=10, validation_data=(X_istiqlal_test, y_istiqlal_test))

model_overall = build_model()
model_overall.fit(X_overall_train, y_overall_train, batch_size=32, epochs=10, validation_data=(X_overall_test, y_overall_test))

# Make predictions
pred_manggarai = model_manggarai.predict(X_manggarai_test)
pred_istiqlal = model_istiqlal.predict(X_istiqlal_test)
pred_overall = model_overall.predict(X_overall_test)

# Inverse transform the predictions to get the original scale
pred_manggarai = manggarai_bkb_scaler.inverse_transform(pred_manggarai)
y_manggarai_test = manggarai_bkb_scaler.inverse_transform(y_manggarai_test.reshape(-1, 1))

pred_istiqlal = istiqlal_scaler.inverse_transform(pred_istiqlal)
y_istiqlal_test = istiqlal_scaler.inverse_transform(y_istiqlal_test.reshape(-1, 1))

pred_overall = overall_scaler.inverse_transform(pred_overall)
y_overall_test = overall_scaler.inverse_transform(y_overall_test.reshape(-1, 1))

# Create DataFrames for predictions and actual values
df_pred_manggarai = pd.DataFrame({
    'Actual': y_manggarai_test.flatten(),
    'Predicted': pred_manggarai.flatten()
}, index=manggarai_bkb_df.index[-len(y_manggarai_test):])

df_pred_istiqlal = pd.DataFrame({
    'Actual': y_istiqlal_test.flatten(),
    'Predicted': pred_istiqlal.flatten()
}, index=istiqlal_df.index[-len(y_istiqlal_test):])

df_pred_overall = pd.DataFrame({
    'Actual': y_overall_test.flatten(),
    'Predicted': pred_overall.flatten()
}, index=overall_df.index[-len(y_overall_test):])

# Display the DataFrames
print("Manggarai BKB Predictions:")
print(df_pred_manggarai.head(24))

print("\nIstiqlal Predictions:")
print(df_pred_istiqlal.head(24))

print("\nOverall Predictions:")
print(df_pred_overall.head(24))



```
